```python
import numpy as np
import pandas as pd

def is_leap_year(year):
    """
    윤년 여부를 판단합니다.
    """
    return year % 4 == 0 and (year % 100 != 0 or year % 400 == 0)

def adjust_values_by_time_and_season(hour, day_of_year, year):

    # 필리핀의 건기와 우기 구분
    is_wet_season = 5 <= (day_of_year // 30) + 1 <= 10  

    # 윤년 처리 
    if is_leap_year(year) and day_of_year > 59:  
        day_of_year -= 1

    peak_hour = 12  # 정오가 최대 발전량 시간

    # 태양의 고도에 따른 발전량 조정 
    solar_intensity = np.cos(np.pi * (hour - peak_hour) / 12)

    # 계절에 따른 발전량 조정
    if is_wet_season:
        solar_intensity *= 0.8  # 우기에는 구름과 강수로 인해 발전량 감소

    return max(solar_intensity, 0)

def generate_solar_data_with_climate_adjustments(start_date, end_date):
    data_list = []
    num_days = (end_date - start_date).days + 1
    base_temperature = 25  # 필리핀의 평균 기온
    base_humidity = 70  # 필리핀의 평균 습도

    for day in range(num_days):
        date = start_date + pd.Timedelta(days=day)
        day_of_year = date.timetuple().tm_yday
        
        for hour in range(24):
            try:
                datetime = pd.Timestamp(date.year, date.month, date.day, hour)
                temperature_adjustment = np.random.normal(0, 5)  # 온도 변동성
                humidity_adjustment = np.random.normal(0, 10)  # 습도 변동성
                temperature = base_temperature + temperature_adjustment
                humidity = base_humidity + humidity_adjustment
                rainfall = np.random.choice([0, 0, 0, 1, 2, 5, 10, 20, 50], p=[0.6, 0.2, 0.1, 0.04, 0.02, 0.02, 0.01, 0.005, 0.005]) if adjust_values_by_time_and_season(hour, day_of_year, date.year) < 0.5 else 0
                solar_intensity = adjust_values_by_time_and_season(hour, day_of_year, date.year)
                power = solar_intensity * 1000  # 가정: 최대 일조량에서 1kW/m²의 발전량을 가짐
                
                data_list.append({
                    "datetime": datetime,
                    "temperature": temperature,
                    "humidity": humidity,
                    "rainfall": rainfall,
                    "solar_intensity": solar_intensity,
                    "power": power,
                })
            except Exception as e:
                print(f"Error generating data for {datetime}: {e}")

    return pd.DataFrame(data_list)

# 데이터 생성 범위 설정
start_date = pd.to_datetime("2023-01-01")
end_date = pd.to_datetime("2023-12-31")

# 데이터 생성
solar_data_adjusted = generate_solar_data_with_climate_adjustments(start_date, end_date)

solar_data_adjusted.head()
```

```python
import numpy as np
import pandas as pd

def adjust_values_by_time_and_season(hour, day_of_year):
    """
    시간과 계절에 따라 발전량을 조정합니다.
    필리핀의 위치와 기후 특성을 고려하여 일조량과 발전량의 변화를 모델링합니다.
    """
    is_wet_season = 5 <= (day_of_year // 30) + 1 <= 10  # 5월부터 10월까지를 우기로 가정
    peak_solar_hour = 10  # 최대 일조량 시간은 오전 10시로 가정 (필리핀의 위치를 고려)
    
    # 시간에 따른 일조량 조정
    solar_intensity = np.cos(np.pi * (hour - peak_solar_hour) / 12)  # 낮 시간대에 최대치 도달
    if is_wet_season:
        solar_intensity *= 0.75  # 우기에는 일조량 감소를 모델링
    
    # 태풍 기간 동안 발전량 추가 조정
    if 150 <= day_of_year <= 300:  # 6월부터 10월까지 태풍이 빈번하게 발생
        solar_intensity *= np.random.uniform(0.4, 0.8)  # 태풍으로 인한 일조량 감소 반영
    
    return max(solar_intensity, 0)

def determine_defect(temperature, humidity, maintenance, age, solar_intensity):
    """
    기후, 유지 보수 상태, 패널 나이, 일조량을 고려하여 결함 여부를 결정합니다.
    """
    # 결함 발생 기본 확률
    defect_probability = 0.01  # 기본적으로 1%의 확률로 결함 발생
    
    # 환경 요인에 따른 확률 조정
    if temperature > 30 or humidity > 80:
        defect_probability += 0.03  # 필리핀의 고온 다습한 환경을 반영하여 결함 확률 증가
    
    # 유지 보수 상태에 따른 확률 조정
    if maintenance == 'Poor':
        defect_probability += 0.05
    elif maintenance == 'Average':
        defect_probability += 0.03
    
    # 패널 나이에 따른 확률 조정
    if age > 5:
        defect_probability += 0.04  # 필리핀의 강한 일조량과 높은 온도로 인한 조기 노화 반영
    
    # 일조량이 매우 낮은 경우 (예: 태풍 기간) 결함 확률 증가
    if solar_intensity < 0.2:
        defect_probability += 0.05
    
    # 최종 결함 여부 결정
    return np.random.rand() < defect_probability

def generate_detailed_solar_data_with_adjustments_and_defects(start_date, end_date):
    data_list = []
    num_days = (end_date - start_date).days + 1

    for day in range(num_days):
        date = start_date + pd.Timedelta(days=day)
        for hour in range(24):
            datetime = pd.Timestamp(date.year, date.month, date.day, hour)
            day_of_year = date.timetuple().tm_yday
            
            # 기본 기후 변수 설정
            temperature = np.random.normal(27, 5)  # 필리핀 평균 온도를 반영
            humidity = np.random.normal(80, 10)  # 필리핀 평균 습도를 반영
            rainfall = np.random.choice([0, 0, 0, 10, 20, 30, 50], p=[0.5, 0.2, 0.1, 0.1, 0.05, 0.025, 0.025])  # 우기에 집중된 강수량 반영
            
            # 발전량 및 일조량 조정
            solar_intensity = adjust_values_by_time_and_season(hour, day_of_year)
            power = solar_intensity * 1000  # 최대 일조량에서의 발전량을 가정
            
            # 유지 보수 상태 및 패널 나이
            maintenance = np.random.choice(['Good', 'Average', 'Poor'], p=[0.5, 0.4, 0.1])  # 유지보수 상태 분포 조정
            age = np.random.choice(range(1, 16))  # 필리핀의 평균 패널 수명을 고려
            
            # 결함 여부 결정
            defect = determine_defect(temperature, humidity, maintenance, age, solar_intensity)
            
            data_list.append({
                "datetime": datetime,
                "temperature": temperature,
                "humidity": humidity,
                "rainfall": rainfall,
                "solar_intensity": solar_intensity,
                "power": power,
                "defect": defect,
                "maintenance": maintenance,
                "age": age
            })
    
    return pd.DataFrame(data_list)

# 날짜 범위 설정 및 데이터 생성
start_date = pd.to_datetime("2010-01-01")
end_date = pd.to_datetime("2024-02-26")
solar_data_with_defects = generate_detailed_solar_data_with_adjustments_and_defects(start_date, end_date)

# 샘플 데이터 확인
solar_data_with_defects.head()

solar_data_with_defects.to_csv("C:/Users/typar/OneDrive/바탕 화면/개인/방송대/태양광/pre_solar_data.csv", index=False)

```
